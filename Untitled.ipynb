```python
%matplotlib notebook
%pylab inline

import numpy
```

```python
from math import sqrt, pi

def total_drag(v):
    return 0.5 * 1.2 * (v**2) * 0.3 * (0.0156 + ((0.6**2) / (pi * 5.63 * 0.8)) + 0.0147)

def velocity(m):
    return numpy.sqrt((2 * m * 9.81) / (0.6 * 1.2 * 0.3))

def distance(drag, weights):
    efficiency = 0.5
    watt_hours_per_kg = 3.6 * 3.2 / 49 * 1000
    weights = weights - 1.0
    
    return (weights * watt_hours_per_kg * 3600) / drag * efficiency

def distance2(drag):
    efficiency = 0.5
    watt_hours_per_kg = 3.6 * 3.2 / 49 * 1000
    weights = 1
    
    return (weights * watt_hours_per_kg * 3600) / drag * efficiency

#print(velocity(2))
#print(total_drag(velocity(2)))
#print(distance(total_drag(velocity(2))))
```

```python
weights = linspace(2, 3)

def foo(weights):
    return distance(total_drag(velocity(weights)), weights)

plot(weights, foo(weights))

xlabel("Weight in Kg")
ylabel("Distance in m")
grid()
```

```python
weights = linspace(1.8, 2)

def foo(weights):
    return distance2(total_drag(velocity(weights)))

plot(weights, foo(weights))

xlabel("Weight in Kg")
ylabel("Distance in m")
grid()
```
